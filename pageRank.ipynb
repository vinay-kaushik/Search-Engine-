```python
import pickle
import time
import os
from copy import deepcopy
start_time = time.time()
number_of_pages = 4000
page_rank = {}
with open("./uic_crawledLinks_"+str(number_of_pages),"rb") as data_file:
    crawled_links = pickle.load(data_file)
    crawled_links_reverse = pickle.load(data_file)

with open("./index_"+str(number_of_pages),"rb") as data_file:
    nodes_inlink = pickle.load(data_file)
    nodes_outlink = pickle.load(data_file)
    inv_index = pickle.load(data_file)
    IDF = pickle.load(data_file)

nodes_score={}
damping_factor = 0.85
number_of_nodes = len(crawled_links.keys())

print(number_of_nodes)
```

```python

def sigma_formula(current_node):
    score = 0
    for neighbour,value in nodes_inlink[current_node].items():
        numerator = nodes_score[neighbour]
        denominator = nodes_outlink[neighbour].keys()
        denominator = len(denominator)
        score += (numerator/denominator)
    return score

def validation():
    sum = 0
    for i,value in nodes_score.items():
        sum = sum + (value)
```

```python
init_score = 1/number_of_nodes
for index,value in crawled_links.items():
    nodes_score[index] = init_score

no_outlinks = []
for index,value in nodes_outlink.items():
    if(value == {}):
        no_outlinks.append(index)

temp_score = deepcopy(nodes_score)
iterations = 30
for i in range(iterations):
    validation()
    dp = 0
    for node in no_outlinks:
        dp += (damping_factor * (nodes_score[node]/number_of_nodes))
    for index in crawled_links.keys():
        temp_score[index] = (damping_factor * sigma_formula(index)) + ((1-damping_factor)/number_of_nodes) + dp
    nodes_score = deepcopy(temp_score)
```

```python
page_rank_order = sorted(nodes_score.items(), key=lambda kv: kv[1], reverse=True)

rank = 1
for page in page_rank_order:
    page_rank[page[0]] = rank
    rank = rank + 1
with open("./pagerank_"+str(number_of_pages),"wb") as output:
    pickle.dump(page_rank, output)

print("--- %s seconds ---" % (time.time() - start_time))

print("PageRank program executed. Please find the pickle file in the mentioned directory")
```
